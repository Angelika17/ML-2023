```python
import pandas as pd
import numpy as np
```

```python
url="https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
syte=pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України")[0]
syte
```

```python
#Task3
syte_1=syte.replace(to_replace="—", value=np.NaN)
print(syte_1.dtypes)
syte_1
```

```python
data= syte_1.apply(pd.to_numeric, errors="ignore")
data
```

```python
len(data.isnull())
```

```python
data.sum()
```

```python
data=data[:-1]
data
```

```python

data.fillna(value=data.mean(),inplace=True )

data=data.round(2)
```

```python
data
```

```python
data[data["2019"]>data["2019"].mean()][["Регіон"]]
```

```python
data[data["2014"]==data["2014"].max()][["Регіон"]]
```
