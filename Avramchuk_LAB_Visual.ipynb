```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
apart.drop(columns=["street","publish_date"])
```

```python
for prop in apart.columns:
    apart[prop] = pd.to_numeric(apart[prop], errors='coerce')
```

```python
apart.head()
```

```python
apart.info()
apart.describe()
```

```python
plt.bar(apart.columns, apart.count())
plt.ylabel("кількість заповнених значень")
plt.xticks(rotation='vertical')
plt.show()
```

```python
missing_data=apart.isna()
missing_data.head()
```

```python

m= missing_data.sum(axis=1)/len(apart.columns)
len(m)
```

```python
apart.dropna(thresh=5, inplace=True)
```

```python
apart.info()
```

```python
apart.fillna(apart.median(), axis=0, inplace=True)
apart.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(apart['area_kitchen'], '.')
plt.title('пошук аномалій в показчику')
plt.xlabel('рядки датасету')
plt.ylabel('м2')
plt.show()
```

```python
fig, axs = plt.subplots(3,3, figsize=(15,15))

for ax, name in list(zip(axs.reshape(9,), apart.columns)):
    ax.set_title(name)

axs[0,0].plot(apart['Price'], '.')
axs[0,1].plot(apart['rooms'], '.')
axs[0,2].plot(apart['price_per_m2'], '.')
axs[1,0].plot(apart['level'], '.')
axs[1,1].plot(apart['levels'], '.')
axs[1,2].plot(apart['year'], '.')
axs[2,0].plot(apart['area_total'], '.')
axs[2,1].plot(apart['area_living'], '.')
axs[2,2].plot(apart['area_kitchen'], '.')
```
